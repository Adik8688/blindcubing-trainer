```python
import pandas as pd
from pathlib import Path
import os

OUT_PATH = Path().absolute().parent / "exports"
filelist = os.listdir(OUT_PATH)
filepath = OUT_PATH / filelist[-1]
```

```python

df = pd.read_excel(filepath)
df.shape
```

### Most unstable algs

```python
df.sort_values('Skew', ascending=True)[:10]
```

### Most stable algs

```python
df.sort_values('Skew', ascending=False)[:10]
```

### Slowest algs

```python
df.sort_values('Mean', ascending=False)[:10]
```

### Fastest algs
