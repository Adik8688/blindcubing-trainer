```python
import pandas as pd
from pathlib import Path
import os

OUT_PATH = Path().absolute().parent / "exports"
filelist = os.listdir(OUT_PATH)
filepath = OUT_PATH / filelist[-1]
```

```python

df = pd.read_excel(filepath)
df.shape
```

### Most unstable algs

```python
df.sort_values('Skew', ascending=False)[:10]
```

### Most stable algs

```python
df.sort_values('Std', ascending=True)[:10][['Buffer', '1st target', '2nd target', 'Alg', 'Count', 'Std']]
```

### Slowest algs

```python
df.sort_values('Mean', ascending=False)[:10][['Buffer', '1st target', '2nd target', 'Alg', 'Count', 'Mean']]
```

### Fastest algs

```python
df.sort_values('Mean', ascending=True)[:10]
```
